```python
aapl= dr.data.get_data_yahoo('AAPL',start='2012-01-01', end='2017-01-01')
cisco= dr.data.get_data_yahoo('CSCO',start='2012-01-01', end='2017-01-01')
ibm= dr.data.get_data_yahoo('IBM',start='2012-01-01', end='2017-01-01')
amzn= dr.data.get_data_yahoo('AMZN',start='2012-01-01', end='2017-01-01')
```

```python
for stock_df in (aapl, cisco, ibm, amzn):
    stock_df['Normed Return']= stock_df['Adj Close']/stock_df.iloc[0]['Adj Close']
```

```python
aapl.head()
```

```python
#30% in apple
#20% in cisco
#40% in amazon
#10% in ibm
```

```python
for stock_df, allo in zip((aapl, cisco, ibm, amzn),[0.3,0.2,0.4,0.1]):
    stock_df['allocation']= stock_df['Normed Return']*allo
    
```

```python
aapl
```

```python
ibm
```

```python
for stock_df in (aapl, cisco, amzn, ibm):
    stock_df['Position Values']= stock_df['allocation']*1000000
```

```python
amzn['Position Values']
```

```python
all_pos_values=[aapl['Position Values'],ibm['Position Values'],cisco['Position Values'],amzn['Position Values']]
```

```python
portfolio_val=pd.concat(all_pos_values,axis=1)
```

```python
portfolio_val.head()
```

```python
portfolio_val.columns=["Apple Position","IBM Position","Cisco Position","Amazon Position"]
```

```python
portfolio_val.head()
```

```python
portfolio_val['Total Position']=portfolio_val.sum(axis=1)
```

```python
portfolio_val.tail()
```

```python
portfolio_val['Total Position'].plot(figsize=(16,8))
```

```python
portfolio_val.drop('Total Position', axis=1).plot(figsize=(16,8))
```

```python
portfolio_val
```

```python
portfolio_val['Daily Returns']= portfolio_val['Total Position'].pct_change(1)
```

```python
portfolio_val
```

```python
portfolio_val['Daily Returns'].plot(kind='kde', figsize=(16,6))
```

```python
cumulative_return= 100* (portfolio_val['Total Position'][-1]/portfolio_val['Total Position'][0]-1)
```
